# Annotate the data according to our manual work

```python
import scanpy as sc
import pandas as pd
from tqdm import tqdm
```

```python
DATA_PATH = "/placodes/data"
```

```python
adata_conc = sc.read_h5ad(f"{DATA_PATH}/anndatas/cranial_placodal_integrated_clonal_clusters.h5ad")
```

```python
annotations=pd.read_csv(f"{DATA_PATH}/annotations/Reannotation placodal and NCC - new_large_set.csv")
```

```python
# Give  annotation
mapping_dict = annotations.set_index('leiden_2')['annotation'].to_dict()
map_dict={}
for x in mapping_dict:
    map_dict[str(x)] = mapping_dict[x]
del mapping_dict
adata_conc.obs['annotation'] = adata_conc.obs['leiden_2'].map(map_dict)
```

```python
# Give broad annotation
mapping_dict = annotations.set_index('leiden_2')['broad_annotation'].to_dict()
map_dict={}
for x in mapping_dict:
    map_dict[str(x)] = mapping_dict[x]
del mapping_dict
adata_conc.obs['broad_annotation'] = adata_conc.obs['leiden_2'].map(map_dict)
```

```python
# Save 
adata_conc.write_h5ad(f"{DATA_PATH}/anndatas/cranial_placodal_integrated_clonal_clusters_annotated.h5ad")
```
